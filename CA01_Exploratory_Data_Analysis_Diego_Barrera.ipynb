# Data Preprocessing

```python
df_preprocessed = df.copy()
```

```python
# Impute missing values with 'median'

#Missing values with over 60%
  #Alley
  #FireplaceQu
  #PoolQC
  #Fence
  #MiscFeature

df_preprocessed.dropna(inplace=True)
print(df_preprocessed.isnull().sum())
```

```python
df_no_missing_value = df_preprocessed.copy()
df_preprocessed_again = pd.get_dummies(df_no_missing_value, columns=['Alley', 'Fence'], prefix=['Alley', 'Fence'])
```

```python
# standardize the values of a numeric column
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_preprocessed_again['YearBuilt_scaled'] = scaler.fit_transform(df_no_missing_value[['YearBuilt']])
print(df_preprocessed_again['_scaled'].head())
```

```python
# bin the values of a numeric column
from sklearn.preprocessing import KBinsDiscretizer

discretizer = KBinsDiscretizer(n_bins=3, encode='ordinal')
df_preprocessed_again['YearBuilt_binned'] = discretizer.fit_transform(df_no_missing_value[['YearBuilt']])
print(df_preprocessed_again.head())
```

```python
# apply a min-max scaling to a numeric column
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_preprocessed_again['YearBuilt_minmax'] = scaler.fit_transform(df_no_missing_value[['YearBuilt']])
print(df_preprocessed_again.head())
```
